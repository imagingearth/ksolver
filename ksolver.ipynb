Regularised kernels solver

Fits the model of reflectance $\rho$:

$$
\hat\rho = f_0 \left[ 1 + f_1' k_1(\Omega,\Omega') + f_2' k_2(\Omega,\Omega') \right]
$$

with $f_1' = f_1/f_0$ and $f_2' = f_2/f_0$ constant over time and $f_0$ regularised (using GCV to determine smoothness).

One core code is a fast smoother with regularisation from Garcia (2010) implemented in Python as the function [`smoothn`](files/smoothn.py).

Another is the kernel BRDF model code in the class [Kernels](files/kernels.py).


Garcia, D. (2010), Robust smoothing of gridded data in one and higher dimensions with missing values, Computational Statistics and Data Analysis 54 (2010) 1167–1178

## Solve

```python
# this will take a few minutes as its solving over a 
# datacube of around 16 x 10^6 samples
k = ksolve(d,drmse_thresh=0.05,verbose=True,doPlot=True)
```

```python
k.plot_t(r=nr/2,c=nc/2,b=1)
```

```python
# showing all results for this
# pixel over all bands
k.plot_s(r=nr/2,c=nc/2)
```

```python
plt.plot(k.d['rmse'])
plt.ylabel('rmse')
plt.xlabel('iteration')
```

```python
# look at f0
plt.figure(figsize=(8,8))
plt.imshow(k.f0[nbands/2,ndays/2],interpolation='nearest')
plt.colorbar()
plt.title('f0 band %d day sample %d'%(nbands/2,ndays/2))
```

```python
# look at f1
plt.figure(figsize=(8,8))
plt.imshow(k.P[0][1],interpolation='nearest')
plt.colorbar()
plt.title("f1' band %d day sample %d"%(1,ndays/2))
```

```python
# look at f2
plt.figure(figsize=(8,8))
plt.imshow(k.P[1][1],interpolation='nearest')
plt.colorbar()
plt.title("f2' band %d day sample %d"%(1,ndays/2))
```

```python
import os

vmin = 0.0
vmax = k.f0.max()

root = 'files/images/f0'
print ndays
for i in xrange(ndays):
    fig = plt.figure(figsize=(8,8))
    plt.imshow(k.f0[1,i],interpolation='nearest',vmax=vmax,vmin=vmin)
    # plot a jpg
    plt.title('%03d'%i)
    plt.colorbar()
    plt.savefig('%s_%s.jpg'%(root,'%03d'%i))
    plt.close(fig)
```
