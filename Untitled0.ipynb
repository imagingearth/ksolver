```python
# now model the residuals
mask = (~Ross_no_fire.mask)
N = np.sum(np.sum(mask,axis=1),axis=0)
k11 = np.sum(np.sum(Ross_no_fire * Ross_no_fire,axis=1),axis=0)/N
k22 = np.sum(np.sum(Li_no_fire * Li_no_fire,axis=1),axis=0)/N
k12 = np.sum(np.sum(Li_no_fire * Li_no_fire,axis=1),axis=0)/N
det = k11*k22 - k12*k12
plt.imshow(det,interpolation='nearest')
plt.colorbar()

f1 = ma.array(np.zeros_like(y_smooth[:,0,:,:]))
f2 = ma.array(np.zeros_like(y_smooth[:,0,:,:]))
# for each band
for i in xrange(fk.shape[0]):
    fkk1 = fkk2 = 0.
    # for each sample
    for j in xrange(Ross_no_fire.shape[0]):
        fkk1 += np.sum(fk[i,j] * Ross_no_fire[j],axis=0)
        fkk2 += np.sum(fk[i,j] * Li_no_fire[j],axis=0)
    fkk1 /= N
    fkk2 /= N
    f1[i] = (fkk1 * k22 - fkk2 * k12)/det
    f2[i] = (fkk2 * k11 - fkk1 * k12)/det
```
